```python
import pandas as pd

df = pd.read_csv("General Data 2014-2020.csv")
df.head()
```

```python
# Check shape and column names
print(df.shape)
print(df.columns)

# Quick overview
df.info()

# Check missing values
print(df.isnull().sum())
```

```python
# Clean object columns
for col in ['ipwirwta', 'ipwirwtr', 'ipaprirta', 'ipaprirtr']:
    df[col] = df[col].replace({',':''}, regex=True)   # remove commas
    df[col] = pd.to_numeric(df[col], errors='coerce') # convert to float
```

```python
print(df.dtypes)   # check if all are numeric now
print(df.head())   # first 5 rows
```

```python
df.to_csv("cleaned_general_data.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

states_df = pd.read_csv("Top 10 State Visit.csv")

state_names = []
visitors = []

for i in range(1, 11):
    state_names.extend(states_df[f'top{i}_state'])
    visitors.extend(states_df[f'top{i}_ftv'])

state_visitors_df = pd.DataFrame({
            'State': state_names,
                'Visitors': visitors
                })

state_visitors_df = state_visitors_df.groupby('State').sum().reset_index()

top5_states = state_visitors_df.sort_values(by='Visitors', ascending=False).head(5)
bottom5_states = state_visitors_df.sort_values(by='Visitors', ascending=True).head(5)

plt.figure(figsize=(10,6))
sns.barplot(x='State', y='Visitors', data=top5_states, palette='viridis')
plt.title("Top 5 States by Tourist Arrivals")
plt.ylabel("No. of Tourists")
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x='State', y='Visitors', data=bottom5_states, palette='magma')
plt.title("Bottom 5 States by Tourist Arrivals")
plt.ylabel("No. of Tourists")
plt.show()
```

```python
countries_df = pd.read_csv("Top 10 Country FFA.csv")

country_names = []
visitors = []

for i in range(1, 11):
    country_names.extend(countries_df[f'top{i}_country'])
    visitors.extend(countries_df[f'top{i}_ftas'])

country_visitors_df = pd.DataFrame({
            'Country': country_names,
                'Visitors': visitors
                })

country_visitors_df = country_visitors_df.groupby('Country').sum().reset_index()

top5_countries = country_visitors_df.sort_values(by='Visitors', ascending=False).head(5)
bottom5_countries = country_visitors_df.sort_values(by='Visitors', ascending=True).head(5)

plt.figure(figsize=(10,6))
sns.barplot(x='Country', y='Visitors', data=top5_countries, palette='viridis')
plt.title("Top 5 Countries by Tourist Arrivals")
plt.ylabel("No. of Tourists")
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x='Country', y='Visitors', data=bottom5_countries, palette='magma')
plt.title("Bottom 5 Countries by Tourist Arrivals")
plt.ylabel("No. of Tourists")
plt.show()
```

```python
month_df = pd.read_csv("Month Wise FFA.csv")

month_df = month_df.melt(id_vars=['year'],
                         value_vars=['january','february','march','april','may','june','july',
                                      'august','september','october','november','december'],
                         var_name='Month', value_name='Visitors')

plt.figure(figsize=(12,6))
sns.barplot(x='Month', y='Visitors', data=month_df, palette='coolwarm', ci=None)
plt.title("Month-wise Tourist Arrivals in India")
plt.ylabel("No. of Tourists")
plt.xticks(rotation=45)
plt.show()
```
